### Intersectional Influence: The Impact of Counterstereotypic Exemplars on Persistence and Perceived Similarity Across Multiple Stigmatized Domains

## Background

   Despite progress towards educational equity, Black students still face unique psychological barriers that contribute to the racial achievement gap. Many of these gaps fall into the category of persistence gaps, such as Black students completing bachelor’s degrees less often than White students (New, 2017). One promising avenue to address these persistence gaps explored in the research is counterstereotypic role models–prominent figures who defy negative stereotypes and reveal a version of what is possible for individuals from a marginalized group (Olsson & Martiny, 2018). Exposure to counterstereotypic role models has been associated with a whole host of positive outcomes, such as increased sense of belonging in science, technology, engineering, and mathematics (STEM), academic performance, and academic persistence (Rivera & Benitez, 2016; Rosenthal et al., 2013; Shachnai & Kushnir, 2022).

   Kamala Harris stands out as a particularly notable example of a counterstereotypic role model. With a long career of consistently breaking achievement barriers for women and people of color under her belt, she is well-positioned as someone to look up to. In recent years, Harris became the first Black South Asian woman to be Vice President of the United States and the first to secure the Democratic Presidential nomination. Her level of success is highly visible to the general public. However, throughout her tenure in the public eye, the media has varied significantly in the extent to which they associate her with different aspects of her identity. For example, when Harris was announced as Joe Biden’s running mate in 2020, major news outlets shared headlines announcing “Kamala Harris officially becomes the first Black woman to be a major party’s vice presidential nominee” (Reston & Collinson, 2020), “Kamala Harris makes history as first female vice president” (Camera, 2020), and “Kamala Harris and the rise of Indian-origin politicians in the West” (Tharoor, 2020), each emphasizing different parts of her multifaceted identity. In order to see someone as a role model, we need to perceive them as similar to ourselves (Dasgupta, 2011).These differences in her portrayals may impact how people think about her and, in turn, influence her effectiveness as a role model. Additionally, despite having a successful career, her very public failure in winning the 2024 Presidential election may also impact her role model effectiveness. As such, she presents a unique opportunity to explore the influence of the salience of a role model’s identity and salient failure on role model effectiveness.

   While existing literature has explored the importance of counterstereotypic role models, the current work addresses three important gaps. First, most existing work connecting role models and persistence focuses on long-term measures of persistence. This study shifts the focus to short-term measures of persistence as it is a mechanism that leads to long-term persistence, and a potentially useful intervention point for counterstereotypic role models. Second, this work extends prior literature by focusing on variability in identity-salience within an individual. Most work focuses on single identity contexts, both of the role model and the participant, but this study aims to provide more context on the impact of salient intersectional identities on persistence. Lastly, the consequences of role model failure are not well understood, so this study explores its impact on persistence behaviors. Therefore, the purpose of this study is to examine the impact of two salient identities (race and gender) on persistence and perceived similarity before (pre-election) and after (post-election) a highly publicized failure among Black Americans.

   In order to test this, we recruited 241 Black participants off of Prolific to participate in the study. Participants were randomly assigned to read a short biography about Kamala Harris’ political journey that emphasized some or no aspect of her racial identity (i.e., Black, Multiracial, no racial identity mentioned) and some or no aspect of her gender identity (i.e., woman, no gender identity mentioned). Then participants completed a one-item measure of perceived similarity before moving on to the persistence task. For the persistence task, participants learned how to solve a cryptarithmetic puzzle, the aim of which is to replace a word equation with a correct arithmetic problem. After completing a practice problem, participants attempt another puzzle, one that they did not know was impossible. Their time spent solving was taken as the primary measure of persistence. Lastly, participants completed a demographics questionnaire.



### Variables of Interest

**Subject ID:** prolific_id


**Demographic Measures:**
 - **Gender:** (rel_gender) As reported by participants (options: man, woman, transgender man, transgender woman, genderqueer; there were no genderqueer participants and transgender participants were grouped with cisgender men and women respectively)
 - **Age:** (age.x) As reported by participants (in years)
 - **Political Affiliation:** (politic_aff.x) As reported by participants (options: democrat = 1, republican = 2, independent = 3, other = 4, none = 5, prefer not to say = 6)

**Independent Variables:**
  - **Harris' salient gender identity:** (KH_gender.x) whether the biography participants read emphasized that Harris was a woman or not
  - **Harris' salient racial identity:** (KH_race.x) whether the biography participants read emphasized that Harris was multiracial, Black, or doesn't emphasize anything about her race

**Outcome variables:**
  - **Time point:** (timepoint) whether the data in that row is from pre-election (0) or post-election (1)
  - **Persistence:** (persist_time) the total time (in seconds) participants spent solving the impossible cryptarithmetic puzzle
  - **Pre-election Perceived Similarity:** (per_sim1) collected at time point 1, using the Inclusion of the Other in the Self (IOS) Scale, participants were presented with a series of 7 images showing two circles, one representing themselves and one representing Kamala Harris, with increasing degrees of overlap. Participants were then asked to “Please select the picture below that best describes how much overlap you believe there is between you and Kamala Harris.” (Aron et al., 1992)
  - **Post-election Perceived Similarity:** (per_sim2) collected at time point 2, using the Inclusion of the Other in the Self (IOS) Scale, participants were presented with a series of 7 images showing two circles, one representing themselves and one representing Kamala Harris, with increasing degrees of overlap. Participants were then asked to “Please select the picture below that best describes how much overlap you believe there is between you and Kamala Harris.” (Aron et al., 1992)

### Hypotheses

We have two primary hypotheses:
- Female participants who encounter Harris’s salient gender identity (black and woman) will persist the most at pre-election, but the least at post-election, compared to all other groups.
- Perceived similarity will mediate the relationship between Harris’s salient identities and persistence, such that participants who perceive greater similarity will persist more pre-election.

### 2. Linear Mixed Effects Model

```r
# mixed effects model
lin_model1 = lmer(persist_time ~ timepoint*KH_gender.x*KH_race.x*rel_gender + (1 | prolific_id),
                                      data = kh_long)
summary(lin_model1)

# mixed effects model (controlling for political affiliation)
lin_model2 = lmer(persist_time ~ timepoint*KH_gender.x*KH_race.x*rel_gender + politic_aff.x + (1 | prolific_id),
                                      data = kh_long)
summary(lin_model2)

## model comparison
compare = AIC(lin_model1, lin_model2)
compare
```

In this section, I ran a mixed effects regression because each participant contributes two rows (pre and post) and I expect there to be meaningful differences in baseline persistence across participants. This model accounts for individual differences in baseline persistence. This model showed that there weren't significant differences in persistence across time (Estimate = 16.18, t = 0.23). Additionally, neither aspects of Harris's salient identities or participant gender alone influenced persistence scores. But we do see that the three way interaction (timepoint:KH_gender.xwoman:KH_race.xmultiracial) is approaching significance, so the impact of timepoint on persistence may depend on Harris's presentation as a multiracial woman. The four-way interaction was not statistically significant but based on the estimate, it may suggest a gendered effect on persistence that depends on the salient identities of Harris Estimate = -286.16, t = -1.41). Pre, Post, and Average persistence by condition and participant gender is plotted below. These results did not support our initial hypothesis.

Additionally, I ran the same model controlling for participant political affiliation. It didn't meaningfully impact the pattern of effects in the data. It did have a slightly lower AIC than the initial model, suggesting it might be a slightly better model fit.

### 3. Bootstrapping

```r
set.seed(101)
boots = 1000
boot_diff = numeric(boots)

for(i in 1:boots){
  boot_sample = kh_long[sample(1:nrow(kh_long), replace = TRUE), ]

  # Filter: only female participants and Black + woman condition
  boot_focal = subset(boot_sample, rel_gender == 1 & KH_gender.x == "woman" & KH_race.x == "black")

  # Get mean persistence pre and post for focal group
  pre = mean(boot_focal$persist_time[boot_focal$timepoint == 0], na.rm = TRUE)
  post = mean(boot_focal$persist_time[boot_focal$timepoint == 1], na.rm = TRUE)

  # Save the difference (post - pre)
  boot_diff[i] = post - pre
}

# Compute 95% CI
ci_low = mean(boot_diff) - 1.96 * sd(boot_diff)
ci_up = mean(boot_diff) + 1.96 * sd(boot_diff)

# Output and plot
ci_low
ci_up
hist(boot_diff, xlab = "Post - Pre Persistence Difference",
     main = "Bootstrap Distribution", col = "lightblue", border = "gray")
```

The full linear mixed effects regression didn't support my initial hypothesis but I decided to do a bootstrap so I could better center my focal group of Women in the Black Woman condition. With a confidence interval of [-62.7, -312.3] being morderately narrow and fully negative, this indicates that the observed drop in persistence from pre- to post-election is both statistically robust and consistent across resamples within this subgroup. This demonstrates  that Women in the Black woman condition persisted less from pre to post election. Unlike the regression, this did directly support my first hypothesis. The cell sizes are quite small since there are so many conditions and only 241 participants so the full regression was likely underpowered but this bootstrap was able to be targeted at the group of interest in a way that revealed additional information about the direction and effect in the group of interest.

### 4. Mediation Analysis

```r
# focus on pre-election data
pre_data = kh_long %>%
  filter(timepoint == 0)
pre_data$KH_gender.x <- factor(pre_data$KH_gender.x, levels = c("none", "woman"))
table(pre_data$KH_gender.x)
pre_data$KH_race.x <- factor(pre_data$KH_race.x, levels = c("black", "multiracial", "none"))
levels(pre_data$KH_race.x)
table(pre_data$KH_race.x)

# mod 1: effect of KH identity salience on similarity
mod1 = lm(per_sim1 ~ KH_gender.x * KH_race.x, data = pre_data)

# mod 2: persistence as predicted by both similarity and identity salience
mod2 = lm(persist_time ~ per_sim1 + KH_gender.x * KH_race.x, data = pre_data)

# summarize both
summary(mod1)
summary(mod2)

# mediation analysis
med.mod = mediate(mod1, mod2, treat = "KH_gender.x", mediator = "per_sim1",
                  covariates = list(KH_race.x = "black"), sims = 1000)
summary(med.mod)
```

This section tests the second hypothesis (perceived similarity mediation at the pre-election time point). In order to test this mediation, I first ran two linear regressions the first looking at the relationship between Harris' salient identities and perceived similarity. Her identities didn't signficantly predict perceived similarity. I also ran a second linear regression looking at the relationship between perceived similarity and persistence. Again, the results show that perceived similarity does not significantly predict persistence. From there, I ran a mediation analysis and as you would expect from the two regressions above, there was no significant mediation effect of perceived similarity on the relationship between Harris' salient identities and persistence at the pre-election time point.

## Conclusions

In this project, I tested two hypotheses. The first hypothesis that female participants who encountered Harris’s salient gender identity (black and woman) will persist the most at pre-election, but the least at post-election, compared to all other groups was not suppported by the initial mixed effects regression. The regression showed no main effects or two way interactions but the three way interaction between Harris' race, gender, and participant gender while still insignificant suggested some trends worth exploring further. I used bootstrapping to estimate the distribution of the sample statistic by resampling with replacement from the dataset. This allowed me to focus on my primary interest Women in the Black woman condition. This suggested that women were persisting less from pre to post election when Harris' identities of being a Black woman were salient when compared to all other groups. Lastly, I tested a mediation model where perceived similarity theoretically mediated the relationship between Harris's salient identities and persistence. Perceived similarity did not mediate this relationship in this dataset nor did Harris' identities predict perceived similarity or did perceived similarity predict persistence. Overall, these results demonstrate the complexity of these high-profile counterstereotypic role models as they don't exist in a vaccum making their influence harder to study. We see some evidence that her presentation as a Black woman impacts Black women but further work is needed to unpack the mechansim and the strength of this influence.
